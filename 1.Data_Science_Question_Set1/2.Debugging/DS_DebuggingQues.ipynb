```python
import numpy as np
from scipy.stats import zscore

data = [1, 2, 3, 100, 5, 6, 7, 8, 9]

for i in data:
    print(i)
    # if zscore(data) > 2:
    print(data.index(i))
    if zscore(data)[data.index(i)] > 2:
        print(f"Outlier detected: {i}")
```

```python
from scipy.stats import norm

pdf_val = norm.pdf(2, 0, -1)
pdf_val
```

```python
norm.pdf(x, loc=0, scale=1)
Because the standard deviation is negative.

In norm.pdf(x, mean, std), the standard deviation must be positive.
You used -1, which is not allowed, so SciPy returns NaN.

Use a positive value instead, like 1.
```

```python
from scipy.stats import norm

pdf_val = norm.pdf(2, 0, 1)
pdf_val
```
